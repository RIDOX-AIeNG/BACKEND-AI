```python
numeric_cols = fs.select_dtypes(include=['int64', 'float64'])
numeric_cols.skew().sort_values(ascending=False)
```

```python
fs['qty_this_week'].plot(kind='box')
```

```python
# Number of weeks a customer made at least one purchase
fs['made_purchase'] = fs['qty_this_week'] > 0

customer_freq = fs.groupby('customer_id')['made_purchase'].sum().sort_values(ascending=False)
customer_freq.head()
```

```python
customer_spend = fs.groupby('customer_id')['spend_this_week'].agg(
    total_spend='sum',
    avg_spend='mean'
).sort_values(by='total_spend', ascending=False)

customer_spend
```

```python
recency_df = fs[fs['qty_this_week'] > 0].groupby('customer_id')['week_start'].max()

# Most recent date in dataset
max_date = fs['week_start'].max()

customer_recency = (max_date - recency_df).dt.days.sort_values()
customer_recency.head()
```
