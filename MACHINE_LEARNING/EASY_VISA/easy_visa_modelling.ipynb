```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, confusion_matrix
import warnings
warnings.filterwarnings("ignore")
```

```python
X_train = pd.read_csv("X_train.csv")
y_train = pd.read_csv("y_train.csv")
X_test = pd.read_csv("X_test.csv")
y_test = pd.read_csv("y_test.csv")
```

```python
def modelling(X_train, y_train, X_test, y_test):
    linear_model = LogisticRegression()
    decision_tree = DecisionTreeClassifier()
    random_forest = RandomForestClassifier()
    xgboost = XGBClassifier()

    linear_model_fit = linear_model.fit(X_train, y_train)
    decision_tree_fit = decision_tree.fit(X_train, y_train)
    random_forest_fit = random_forest.fit(X_train, y_train)
    xgboost_fit = xgboost.fit(X_train, y_train)

    linear_model_preds = linear_model_fit.predict(X_test)
    decision_tree_preds = decision_tree_fit.predict(X_test)
    random_forest_fit_preds = random_forest_fit.predict(X_test)
    xgboost_fit_preds = xgboost_fit.predict(X_test)

    print("========Linear Model=========")
    print(f"Accuracy: {accuracy_score(y_test, linear_model_preds)}")
    print(f"f1 score: {f1_score(y_test, linear_model_preds)}")
    print(f"recall score: {recall_score(y_test, linear_model_preds)}")
    print(f"precision score: {precision_score(y_test, linear_model_preds)}")
    print("\n")

    print("========Decision Tree========")
    print(f"Accuracy: {accuracy_score(y_test, decision_tree_preds)}")
    print(f"f1 score: {f1_score(y_test, decision_tree_preds)}")
    print(f"recall score: {recall_score(y_test, decision_tree_preds)}")
    print(f"precision score: {precision_score(y_test, decision_tree_preds)}")
    print("\n")

    print("========Random Forest=========")
    print(f"Accuracy: {accuracy_score(y_test, random_forest_fit_preds)}")
    print(f"f1 score: {f1_score(y_test, random_forest_fit_preds)}")
    print(f"recall score: {recall_score(y_test, random_forest_fit_preds)}")
    print(f"precision score: {precision_score(y_test, random_forest_fit_preds)}")
    print("\n")

    print("=========xgboost========")
    print(f"Accuracy: {accuracy_score(y_test, xgboost_fit_preds)}")
    print(f"f1 score: {f1_score(y_test, xgboost_fit_preds)}")
    print(f"recall score: {recall_score(y_test, xgboost_fit_preds)}")
    print(f"precision score: {precision_score(y_test, xgboost_fit_preds)}")
    print("\n")

modelling(X_train, y_train, X_test, y_test)
```
