**Feature Engineering**

```python
bins = [6000,	28000,	47000,	61000,	290000, float('inf')]
labels = ['Low', 'Medium', 'High', 'Very High', 'High earners']

# Create new column for income category
df['Income_Category'] = pd.cut(df['annual_income'], bins=bins, labels=labels)
df['Income_Category_Code'] = df['Income_Category'].map({'Low': 1, 'Medium': 2, 'High': 3, 'Very High': 4, 'High earners':5})
```

```python
income_category = pd.DataFrame(df["Income_Category"]).value_counts()
print(income_category)
```

```python
num_cols = ["Income_Category", "debt_to_income_ratio","credit_score","loan_amount","interest_rate","loan_paid_back"]
# for i in num_cols:
#         df[i] = np.log(df[i])
```
